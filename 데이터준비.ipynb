```python
def date(st):
    return st[:st.find(':')-3]
def si(st):
    return st[st.find(':')-2:st.find(':')]

df_2106_daeban['대여일'] = df_2106_daeban['대여일시'].apply(date)
df_2106_daeban['대여시'] = df_2106_daeban['대여일시'].apply(si)

df_2106_daeban['반납일'] = df_2106_daeban['반납일시'].apply(date)
df_2106_daeban['반납시'] = df_2106_daeban['반납일시'].apply(si)

df_2106_daeban
```

```python
daeyeo = df_2106_daeban[['자전거번호', '대여 대여소번호', '대여일', '대여시']]
daeyeo.columns = ['자전거번호', '대여소번호', '일', '시']
daeyeo = daeyeo.groupby(by=['대여소번호', '일', '시']).count()
daeyeo.columns = ['시간별대여대수']
daeyeo
```

```python
banap = df_2106_daeban[['자전거번호', '반납대여소번호', '반납일', '반납시']]
banap.columns = ['자전거번호', '대여소번호', '일', '시']
banap = banap.groupby(by=['대여소번호', '일', '시']).count()
banap.columns = ['시간별반납대수']
banap
```

```python
chong = banap.join(daeyeo, how='outer')
chong = chong.fillna(0)
chong
```

```python
chong['총 대수'] = chong['시간별반납대수'] - chong['시간별대여대수']
chong
```

```python
data = pd.DataFrame(list(collection.find()))
data
```

```python
data = data.drop(['_id'], axis = 1)
data
```

```python
data.dtypes
```

```python
train = data.loc[data['일'] <= '2021-06-25']
train = train.set_index(['대여소번호', '시'])
train = train.drop(['일', '시간별반납대수', '시간별대여대수', '설치시기', 'LCD운영대수', 'QR운영대수'], axis = 1)
train = train.reset_index().set_index('대여소번호')
train
```

```python
test = data.loc[data['일'] > '2021-06-25']
test = test.set_index(['대여소번호', '시'])
test = test.drop(['일', '시간별반납대수', '시간별대여대수', '설치시기', 'LCD운영대수', 'QR운영대수'], axis = 1)
test = test.reset_index().set_index('대여소번호')
test
```

```python
target = '총 대수'
features = train.drop(columns=[target]).columns

X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]
```

```python
from category_encoders import OneHotEncoder
from sklearn.feature_selection import f_regression, SelectKBest
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor

pipe = make_pipeline(
    OneHotEncoder(), 
    SimpleImputer(strategy='mean'), 
    RandomForestRegressor(max_depth = 10, n_jobs=-1, random_state=2)
)

k = 20

scores = cross_val_score(pipe, X_train, y_train, cv=k, 
                         scoring='neg_mean_absolute_error')

print(f'Score for {k} folds:', -scores)
-scores.max()
```

```python
from category_encoders import OrdinalEncoder
pipe = make_pipeline(
    OrdinalEncoder(), 
    SimpleImputer(strategy='mean'), 
    RandomForestRegressor(max_depth = 10, n_jobs=-1, random_state=2)
)

k = 20

scores = cross_val_score(pipe, X_train, y_train, cv=k, 
                         scoring='neg_mean_absolute_error')

print(f'Score for {k} folds:', -scores)
-scores.max()
```

```python
from sklearn.metrics import mean_absolute_error

pipe = make_pipeline(
    OneHotEncoder(), 
    SimpleImputer(strategy='mean'), 
    RandomForestRegressor(max_depth = 10, n_jobs=-1, random_state=2)
)
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
mae_test = mean_absolute_error(y_test, y_pred)
print(f'테스트세트 MAE: {mae_test:,.0f}')
```
